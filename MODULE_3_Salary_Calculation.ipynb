Salary Calculation

```python
def calculate_salary(basic_salary, present_days):
    print("\n--- Salary Calculation ---")

    pf = basic_salary * 0.12

    if present_days < 3:
        hra = 0
        da = 0
        print("Attendance less than 3 days â†’ No HRA & DA")
    else:
        hra = basic_salary * 0.20
        da = basic_salary * 0.10

    net_salary = basic_salary + hra + da - pf
    print("Net Salary:", net_salary)

    return net_salary
```

```python
calculate_salary(basic_salary=50000, present_days=25)
```
